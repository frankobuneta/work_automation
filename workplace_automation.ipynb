```python
# Converting object dtypes to more appropriate dtypes
appearances['appearance_id'] = appearances['appearance_id'].astype('string')
appearances['player_name'] = appearances['player_name'].astype('string')
appearances['competition_id'] = appearances['competition_id'].astype('string')
appearances['date'] = pd.to_datetime(appearances['date'])
```

```python
# Some more information about the DataFrame
appearances.info()
```

```python
# Before we convert to category, let's save columns and data types
# This code is a bit advanced stuff

{column_name: f'{column_series.dtype}' for column_name, column_series in appearances.items()}
```

```python
# Total number of rows from shape
tot_rows = appearances.shape[0]

# Loop for every column in the table generating two variables every time: column name and column series (column data)
for column_name, column_series in appearances.items():
    # If unique values of the column are 20% of total rows
    if column_series.nunique() < tot_rows * 0.2:
        # Convert to category
        appearances[column_name] = appearances[column_name].astype('category')
```

```python
# Function to convert all the appropriate columns to category
def convert_to_category(data_frame_tc: pd.DataFrame) -> pd.DataFrame:
    # Total number of rows from shape
    tot_rows = data_frame_tc.shape[0]
    
    # Loop for every column in the table generating two variables every time: column name and column series (column data)
    for column_name, column_series in data_frame_tc.items():
        # If unique values of the column are 20% of total rows
        if column_series.nunique() < tot_rows * 0.2:
            # Convert to category
            data_frame_tc[column_name] = data_frame_tc[column_name].astype('category')
```

```python

is_liverpool = app_with_club['club_name'].fillna('').str.contains('Liverpool')
criteria_2 = (app_with_club['goals'] > 3)
c = is_liverpool & criteria_2

# This is my favourite feature - changing data in rows that only have that criteria
app_with_club.loc[c, 'new_column'] = 'Top goals scored in a game'
```

#### Groupby (Pivot tables)

```python
# Aggregation - reset_index
liverpool = app_with_club[is_liverpool].copy().reset_index(drop=True)

liverpool.groupby('player_name', observed=True).agg({'goals': 'sum'})
```

```python
# Creating a new variable that will contain groupby data
goals_pivot = liverpool.groupby('player_name', observed=True).agg({'goals': 'sum'}).reset_index()
# Reseting index the second time when we are sorting values
goals_pivot.sort_values(by='goals', ascending=False, inplace=True).reset_index(drop=True)
goals_pivot
```

#### Integration with a database

```python
# Creating a database connection, and copying the data to the table GOALS

import sqlite3

db_connection = sqlite3.connect('files/footbal_data.db')
goals_pivot.to_sql('GOALS', db_connection, if_exists='replace')
```

```python
# Getting the goals data to a DataFrame from database
goals_data = pd.read_sql('SELECT * FROM GOALS WHERE goals > 50', db_connection)
goals_data
db_connection.close()
```
